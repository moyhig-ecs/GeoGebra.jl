```julia
using IJulia
k = IJulia.Kernel()
println(k)                           # 表示される情報を確認
println(fieldnames(typeof(k)))       # 構造体のフィールド一覧を表示
# もし .id/.uuid などがあればアクセスしてみる:
if :id in fieldnames(typeof(k))
    println("kernel id via Kernel(): ", getfield(k, :id))
end
```

```julia
using PyCall
ipykernel = pyimport("ipykernel.connect")
cf = ipykernel.get_connection_file()          # e.g. "/.../kernel-<ID>.json"
m = match(r"kernel-(.*)\.json", cf)
kernel_id = m !== nothing ? m.captures[1] : nothing
println("connection file: ", cf)
println("kernel id: ", kernel_id)
```
